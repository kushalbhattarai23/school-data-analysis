### Age Category

```python
data['age']
```

```python
def age_group(student_age):
    if student_age >=15 and student_age<=17:
        return 1
    elif student_age>=18 and student_age<=20:
        return 2
    else:
        return 3
```

```python
data['age_category']=data['age'].apply(age_group)
data['age']
```

```python
data[['age','age_category']]
```

### Passes (yes or no) column

```python
data['G3']
```

```python
def finalresult(marks):
    if marks>=8:
        return "yes"
    else:
        return "no"
    
```

```python
data["passed (yes or no)"]=data['G3'].apply(finalresult)
data['G3']
```

```python
data[['G3',"passed (yes or no)"]]
```

## Initial Data Analysis

### Summary statistics for age, absences, G1, G2 and G3

```python

col_list = ['age','absences','G1','G2','G3']
for each in col_list:
    output = {  'sum': data[each].sum(),
                 'mean': data[each].mean(),
                 'median': data[each].median(),
                 'standard_deviation' : data[each].std(),
                 'max' : data[each].max(),
                 'min' : data[each].min()
             }
    print({each: output})
    print()
```

## Further Data Analysis

### Students pass / fail percentage

### Sex Pass/Fail Percentage


```python
# Count the number of passed/failed for each sex
pass_counts=data.groupby(['sex', 'passed (yes or no)'])['sex'].count()

# Extract pass/fail counts for males and females
male_pass = pass_counts['M']['yes']
male_fail = pass_counts['M']['no']
female_pass = pass_counts['F']['yes']
female_fail = pass_counts['F']['no']

# Create a pie chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot pie chart for males
ax1.pie([male_pass, male_fail], labels=['Passed', 'Failed'], autopct="%.0f%%", radius=1)
ax1.set_title("Males Percentage")

# Plot pie chart for females
ax2.pie([female_pass, female_fail], labels=['Passed', 'Failed'], autopct="%.0f%%", radius=1)
ax2.set_title("Females Percentage")

# Display the chart
plt.show()
```

### Address Analysis

```python
# Group the data by address type and pass/fail status
address_counts = data.groupby(['address', 'passed (yes or no)'])['address'].count()

# Extract the pass/fail counts for urban and rural students
urban_pass = address_counts['U']['yes']
urban_fail = address_counts['U']['no']
rural_pass = address_counts['R']['yes']
rural_fail = address_counts['R']['no']

# Create a bar graph
fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.35
opacity = 0.8

# Plot the data for urban students
urban_pass_bar = ax.bar('Urban', urban_pass, bar_width, alpha=opacity, color='g', label='Passed')
urban_fail_bar = ax.bar('Urban', urban_fail, bar_width, alpha=opacity, color='r', bottom=urban_pass, label='Failed')

# Plot the data for rural students
rural_pass_bar = ax.bar('Rural', rural_pass, bar_width, alpha=opacity, color='g', label='Passed')
rural_fail_bar = ax.bar('Rural', rural_fail, bar_width, alpha=opacity, color='r', bottom=rural_pass, label='Failed')

# Add labels and a legend
ax.set_xlabel('Address Type')
ax.set_ylabel('Count')
ax.set_title('Passed/Failed Students Based on Address')
ax.legend()

# Display the graph
plt.show()
```
